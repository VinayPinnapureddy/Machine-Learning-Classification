```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ds=pd.read_csv("Social_Network_Ads.csv")
ds.head()
```

```python
x=ds.iloc[:,:-1].values
y=ds.iloc[:,-1].values
print(x)
y=y.reshape(len(y),1)
print(y)
```

```python
print(x.shape,y.shape)
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy="mean")
si.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
print(x_train,x_test)
print(y_train,y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
print(x_train)
print(x_test)
```

```python
from sklearn.linear_model import LogisticRegression 
lr=LogisticRegression(solver="lbfgs")
lr.fit(x_train,y_train.ravel())
```

```python
y_pred=lr.predict(sc.transform([[5,878]]))
print(y_pred)
```

```python
y1_pred=lr.predict(x_test)
y1_pred=y1_pred.reshape(len(y1_pred),1)
print(y1_pred)
print(y_test)
print(y1_pred.shape,y_test.shape)
print(np.concatenate((y1_pred.reshape(len(y1_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
print(confusion_matrix(y_test,y1_pred))
```

```python
print(accuracy_score(y1_pred,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
para={"C":[10**-4,10**-2,10**0,10**2,10**4]}
print(type(para))
model=GridSearchCV(LogisticRegression(),para,scoring='f1',cv=5)
model.fit(x_train,y_train.ravel())
print(model.best_estimator_)
print(model.score(x_test,y_test.ravel()))
```

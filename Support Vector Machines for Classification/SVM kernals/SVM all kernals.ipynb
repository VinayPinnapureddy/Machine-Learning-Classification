```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ds=pd.read_csv("Iris.csv")
ds.head()
```

```python
x=ds.iloc[:,1:5].values
y=ds.iloc[:,-1].values
print(x)
y=y.reshape(len(y),1)
print(y)
print(type(y))
```

```python
for i in y:
    print(i)
    print(type(i))
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
y=l.fit_transform(y)
#y=y.reshape(len(y),1)
print(y)
print(type(y))
a=[]
for i in y:
    a.append(int(i))
print(a)
```

```python
a=[]
for i in y:
    a.append(int(i))
print(a)
print(type(a))
b=np.array(a)
print(b)
print(type(b))
b=b.reshape(len(b),1)
print(b)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,b,test_size=0.25,random_state=1)
print(y_train)
y_train=y_train.reshape(len(y_train),1)
print(type(y_train))
print(np.ndim(y_train))
```

```python
def evaluate_on_test_data(model=None):
    predictions=model.predict(x_test)
    correct_classifications=0
    for i in range(len(y_test)):
        if predictions[i]==y_test[i]:
            correct_classifications+=1
    accuracy=100 * (correct_classifications)/len(y_test)
    return accuracy
```

```python
from sklearn import svm
kernels=("linear","poly","rbf")
accuracies=[]
for index,kernel in enumerate(kernels):
    model=svm.SVC(kernel=kernel,gamma="auto")
    model.fit(x_train,y_train.ravel())
    acc = evaluate_on_test_data(model)
    accuracies.append(float(acc))
    print(" {} % accuracy obtained with kernel ={}".format(acc,kernel))
```

```python
svc=svm.SVC(kernel="linear",gamma="auto").fit(x_train,y_train.ravel())
rbf_svc=svm.SVC(kernel="rbf",gamma=0.7).fit(x_train,y_train.ravel())
poly_svc=svm.SVC(kernel="poly",degree=3,gamma="auto").fit(x_train,y_train.ravel())
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
for index,kernel in enumerate(kernels):
    model=svm.SVC(kernel=kernel,gamma="auto")
    model.fit(x_train,y_train.ravel())
    acc=accuracy_score(model.predict(x_train),y_train)
    accuracies.append(float(acc))
    print(" {} % accuracy obtained with kernel ={}".format(100*acc,kernel))
```
